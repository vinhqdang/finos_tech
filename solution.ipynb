# Data Exploration

```python
import pandas as pd
import pandas_profiling
```

```python
df = pd.read_csv("Data.csv")
```

```python
df.head()
```

```python
# Convert the string column to datetime format
df ['ISSUE_DATE'] = pd.to_datetime (df ['ISSUE_DATE'], infer_datetime_format=True)
df ['ST_DATE'] = pd.to_datetime (df ['ST_DATE'], infer_datetime_format=True)
# df ['PROPOSAL_SUBMISSION_DATE'] = pd.to_datetime (df ['PROPOSAL_SUBMISSION_DATE'], infer_datetime_format=True)

# Adding 1 column: number of days from the first issue to updated day
df["DIFF_ST_DATE"] = (df ['ST_DATE'] - df ['ISSUE_DATE']).astype('timedelta64[D]')

# Create new working dataframe
# Remove constant columns and columns of nan
df01 = df.dropna (axis = 1, how = 'all')
df01 = df01.loc[:, (df01 != df.iloc[0]).any()] 
```

```python
# Profiling the dataframe
pandas_profiling.ProfileReport(df01)
```
